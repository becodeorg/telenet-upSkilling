## Learning Objectives
You will learn how to create an API in python with the Flask module.
At the end of the challenge, you will be able to:
- Create a flask app
- Add a route that show hello world
- Add a route with GET method
- Add a route with POST method

## The Mission
Your work for a company that ask you to create and API to do a demo to their new customer.

You decide to use Flask as it's the best tool to boostrap and API in python.

You receive this mail from the client:

```
Hi Dave,
We would like an demo API to see what you're capable of.

We will ask you the basics of APIs.
I attach to this mail the requirements.

Regards,
Aly company
```

#### Attachments:
```
# Route to check if the server runs
GET /status -> "Alive!"

# Route that return the user's firstname and lastname
GET /user/<firstname:str>/<lastname:str> -> "hello! {FIRSTNAME} {LASTNAME}"

# A route that take the username + password and return the password's hash
POST /password 
input: { username: <USERNAME>, password: <PASSWORD> }
output: { username: "<USERNAME>", password: "<PASSWORD_HASH>" }

```


### Must-have features
- A `GET` route at `/status` that return the string "alive".
- A `GET` route at `/user` that return the user's firstname and lastname.
- A `POST` route at `/password` that return a string containing "Login success for user `USER_NAME_HERE` with password's hash. 

### Nice-to-have features
- Folow pep8 rules.
- Add a docstring for each route.
- Type all you functions.
- Use `Black` to format your code.
- Add a route that take an image and save it on the server. Return the image's server path.
- Add unittest for at least one route.


### Miscellanous information
Feel free to play with Flask and add you own routes!
