## Learning Objectives
You will learn how to properly use Docker and use importants concept of it:
* Images
* Containers
* Volumes

At the end of the challenge, you will be able to:
- Create your own Dockerfile.
- Build a Docker image.
- Run a Docker container.
- Add a Docker volume to your container.
- Manage your images
- Manage your containers

## The Mission
You work for a company that create ML models and ship them to their customer using APIs.
Your team leader want to ship the next project using docker. He asks you to understand the basics of Docker in order to be effective for the next project.

**WARNING:** For this challenge you **have to put your dockerfile in an isolate folder nammed docker**. **It's not a best practice** at all (do not do that later).
The goal is to learn how you can build your image from a different folder than the folder where you have you code.

You will reproduce the architecture of your collegues. Here it is:

```
/app
    |-docker
    |   |-Dockerfile -> your Dockerfile
    |-pipeline
    |   |
    |   |-model
    |   |    |-model.py -> print a number between 1 and 400
    |   |-preprocessing
    |   |    |-preprocessing.py -> print a numpy array
    |   |-utils
    |   |    |-utils.py -> print "in progress..."
```

### Steps
1. Create a github repository.
2. Create the above achitecture in it.
3. Create a Dockerfile that copy those files in a `/app` folder and run the `model.py` file.
4. Build your image
5. Run a container and verify that it print the ouput of you `model.py` file.
6. Connect to your container with SSH and runn all the python files `model.py`, `preprocessing.py`, `utils.py`
7. Stop you container and delete it *(check that there is no more container when using `docker container ls -a`)*
8. Run a new container using a volume to do changes without stoping the container and saving your changes. Connect to SSH so it stay running.
9. Add two lines of code in each files locally (be creative!)
10. Re-run all your python files to veirfy that the changes are effective in your container.
11. Delete all your **containers** and **images**.
12. Push your repo with your Dockerfile.

### Must-have features
- A nice commented Dockerfile
- The same file architecture as above
- There no more images or containers in your system

### Nice-to-have features
- Folow the [dockerfiles best practices](https://www.docker.com/blog/intro-guide-to-dockerfile-best-practices/)
